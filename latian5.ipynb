```python
def normal_saphiro(col) :
    from scipy.stats import shapiro
    SStat, SPval = shapiro(le[col])

    if SPval > 0.05 :
        # Distribusi Normal
        print (f"SPvalnya adalah : {SPval}")
        return True 

    else :
        # Tidak Distribusi Normal
        print (f"SPvalnya adalah : {SPval}")
        return False
```

```python
variables = ['Life expectancy ', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B',
       ' BMI ', 'Polio', 'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       'Schooling']

for i in variables:
    nama_var = i
    print(f"\nUntuk Variabel {nama_var}")
    if (normal_saphiro(nama_var)) :
        rata2 = le[nama_var].mean()
        le[nama_var] = le[nama_var].fillna(value=rata2)
        print(f"Missing value sudah diganti dengan Mean. Yaitu adalah {rata2}")
    else :
        median = le[nama_var].describe()["50%"]
        le[nama_var] = le[nama_var].fillna(value=median)
        print(f"Missing value sudah diganti dengan Median. Yaitu adalah {median}")
```
